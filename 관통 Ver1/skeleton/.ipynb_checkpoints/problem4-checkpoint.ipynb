# 문제4. 상품과 옵션 정보들을 담고 있는 새로운 객체를 만들어 반환하시오.
- [힌트] 상품 리스트와 옵션 리스트를 금융상품 코드를 기준으로 매칭할 수 있습니다.
- [힌트] 아래와 같은 순서로 데이터를 출력하며 진행합니다.
- 1. 응답을 json 형식으로 변환합니다.
- 2. key 값이 "result" 인 데이터를 변수에 저장합니다.
- 3. 2번의 결과 중 key 값이 "baseList" 인 데이터를 변수에 저장합니다.
- 4. 2번의 결과 중 key 값이 "optionList" 인 데이터를 변수에 저장합니다.
- 5. 3번에서 저장된 변수를 순회하며, 4번에서 저장된 값들에서 금융 상품 코드가 
-     같은 모든 데이터들을 가져와 새로운 딕셔너리로 저장합니다.
-     저장 시, 명세서에 맞게 출력되도록 저장합니다.
- 6. 5번에서 만든 딕셔너리를 결과 리스트에 추가합니다.

```python
def get_deposit_products():
    # 본인의 API KEY 로 수정합니다.
    api_key = '9a62dd04881380f095aead4ae0b3060f'
    
    # 요구사항에 맞도록 이곳의 코드를 수정합니다.
    url = 'http://finlife.fss.or.kr/finlifeapi/depositProductsSearch.json'
    params = {
        'auth' : api_key,
        'topFinGrpNo' : '020000',
        'pageNo' : 1
    }
    response = requests.get(url, params = params).json()

    return response
```

```python
# 아래 코드는 수정하지 않습니다.
if __name__ == '__main__':
    # json 형태의 데이터 반환
    result = get_deposit_products()
    # prrint.prrint(): json 을 보기 좋은 형식으로 출력
    #pprint.pprint(result)
```

```python
baselist = result['result']['baseList']
```

```python
optionlist = result['result']['optionList']
```

### 정답 코드

```python
result_list = [] # 마지막에 결과를 담을 리스트
for i in baselist: # 상품 리스트를 상위로
    options = [] # 코드가 일치하는 옵션들을 담을 빈 리스트 생성. 옵션리스트 내의 반복이 끝나면 초기화되도록! 
    for k in optionlist: # 옵션리스트를 중첩 반복
        if i['fin_prdt_cd'] == k['fin_prdt_cd']: # 상품 코드가 일치하는 경우 비교
            option_data = {'저축 금리' : k['intr_rate'], '저축 기간' : k['save_trm'], '저축금리유형' : k['intr_rate_type'],
                                '저축금리유형명' : k['intr_rate_type_nm'], '최고 우대금리' : k['intr_rate2']} # 필요한 option들만 추출해서 option_data 딕셔너리 생성
            options.append(option_data) # options 리스트에 동일한 상품 코드의 옵션 딕셔너리들을 추가
    product_data = {'금리정보': options, '금융상품명': i['fin_prdt_nm'], '금융회사명': i['kor_co_nm']} # 금융상품에 대해 문제 명세서에 맞는 형식으로 딕셔너리 생성

    result_list.append(product_data) # baselist 한 번 반복할 때마다, 결과 리스트에 추가
```

```python
print(result_list)
```
